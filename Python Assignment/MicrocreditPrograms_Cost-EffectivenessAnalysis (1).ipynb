### I. Introduction

In the modern economic world today, most of the poor people, especially in developing economies, do not have access to any financial services that provided by formal lending. They are exposed to risk of being credit constained. Therefore, a large proportion of the poor population must rely on informal lending such as loand from family or/and friends, or from individual lenders that lend them at higher interest rate to ensure that the loans will be repaid. Microfinance is a type of banking service that provides substantial flows or credit, often to low-income individuals and households who otherwise have no access to financial services. Microfinance is more than simply lending and collecting. These programs can serve as an instrument to alleviate poverty and promote economic growth and equality worldwide. Several microfinance Monetary Financial Institutions (MFIs) have established and have been operating towards resolving the credit access problem of the poor.
 
If the microcredit institutions can improve the access to credit of the poor, then these people can finance productive activities to increase their income. Borrowing by poor people is a good solution to improve the redistribution of income if the economy has little growth (Khandker R.S., 2003). These programs are very important to the poor because they can increase per capita consumption and help them acquire the basic necessities for their families. Additionally, microcredit programs, promote  investment in human capital such as schooling. This topic has drwan the attention of many economists and generated a large body of studies that investigated the connection between the microfinance and poverty alleviation.

### II.  Microfinance Performance and Cost-Effectiveness Analysis : Literature Review

Worldwide, microfinance institutions (MFIs) expand the frontier of finance by providing loans and other financial services to under-served populations (Hartarska V. and Nadolnyak D., 2008). A cost-effectiveness analysis is essential to examine how the different implementations by microfinance institutions  compare in efficiency terms with other means of transferring income to the poor. It is true that there is a general expecation that the Monetary Financial Institutions (MFIs) are effective to transfer the income and provide sustainable services to the low-income people. In addition,it is significant to look in detail and judge the effectivenes of different alternatives as well. A cost-effectiveness analysiss is a topic that few studies have studied. 

Shahidur R. Khandker (1998) state that microfinance programs are effective to reduce the poverty among the poor people, particularly among women. They collected survey data from Bangladesh. Bangladesh is very poor country, but a pioneer of the microfinance movement. Many people in Bangladesh do not have access to financial services, especially women. The author examines the cost-effectiveness of microfinance in Bangladesh as compared with other financial measures from other institutions that attempt to reduce the poverty.

The survey by Albert Park and Changging Ren (2001) analyzes household survey data collected in 1997 from three Chinese microcredit programs. These Chinese programs are based on the model og Grameen Bank in Bangladesh. The author made within-country comparisons of NGO programs, mixed programs and government ownerships. They found that the performance of the governmental programs is very poor in contrast with the other types of non-governmental programs.
  
The study by Steven B. Caudill, Daniel M. Gropper and Valentina Hartarska (2009) points out that the Monetary Fiancial Institutions are vital in many developing economies. They collected data from MFIs in Central Asia and Eastern Europe and they found that the longer an MFI is in operation the lower the operation costs are. The cost reductions are related to several factors. Additionally, the larger MFIs that offer deposits and those MFIs that receive lower subsidies can operate more cost-effectively over time.  
    
Furthermore, the article by Paul Mosley (2001) attempts to make a cost-effectiveness analysis of the Monetary Financial Institutions relative to other anti-poverty policy measures and governmental interventions in Bolivia. This survey examines the difference in institutional performance between four different microfinance institutions. The author suggest that microfinance intervention is relatively cheaper and more succesful than the other meausures, but less effective compared to infrastructural meausures in reducing extreme poverty. In addition, microfinance transfer is not dominated by any single model.
  
The results suggest that subsidies to Grameen Bank are more cost-effective ways to transfer the income to the poor people relative to other programs that provided by other formal financial institutions such as the BRAC Bank, RAKUB Bank etc. Women who decided to participate in microfinance programs improved their welfare level. The author mentions that considering the opportunity cost of not using the funds in other development activities is another way of looking at costs.

The survey by Befekadu B.Kereta (2007) attempted to look at Monetary Fiancial Institutions' performance in Ethipoia from outreach and financial sustainability angles using data obtained from primary and secondary sources. Ethiopia is one of the least developed countries. The author suggest that the Monetary Financial Institutions in Ethipoia are profitable. Furthermore, women access is still limited in the finance programs provided by formal lending. Therefore, women's access to credit has to be strengthened. Finally, the empirical evidence of the article indicates a positive correlation between outreach and financial sustainability implies that we could reach more client to attain social mission and as well we could be profitable.

The article by Karel Janda and Pavel Zetek (2014) worths a careful look because it examines the impact of public expenditures, and government debt as well, in microfinance performance. The authors collected data from Monetary Financial Institutions in Latin America and Caribbean. Specifically, the authors suggest, that Monetary Financial Institutions operating in the country with higher growth of Gross Domestic Product (GDP) are characterized by higher rate of social efficiency. In addition, the positive influence on microfinance is also associated with a growth of rural population or an economy openness of the given country.

The paper by Hartarska V. and Nadolnyak D. (2008) uses data collected from the Living Standards Measurement Survey of the World Bank, the Local Initiative Project and the Association of Bosnian Microfinance Institutions to examine the microfinance performance for microenterprises in Bosnia and Herzegovina. The authors found that microenterprises with improved assess to credit rely less on internal funds for their investments. Moreover, the results indicate that the microfinance programs minimize or/and eliminate the financial constraints of the microenterprises.

Finally, the syrvey by Babajide Abiola (2011) examines the microfinance sector in Nigeria. More precisely, the author studies whether microfinance institutions improved access to credit for microenterprises in Nigeria or not. It uses data collected from a field survey and Central Bank of Nigeria data on Microfinance Banks in Nigeria by location. The author found that Microfinance Banks improved access to credit in locations where more Microfinance Banks offered financial products. The reason for that is that investment in local microenterprises was less sensitive to availability of internal funds in unconstrained location, than investment in microenterprises in locations where microfinance activities were limited or non-existent and where micro entrepreneurs had to rely more on internal funds for investment.

### III. Methodology - Approach

In order to create a cost-effectiveness analysis, one eneds to estimate costs and output. The cost-effectiveness of a project, in this case a microfinance program, can be estimated by dividing the two (cost per unit of output). We are explicitly interested in the social costs of such a program, not the private costs. Schreiner (2003) for example, in investigating the cost-effectiveness of the Grameen Bank  takes the number of memberships as output. The cost-effectiveness analysis is therefore calculated as the discounted subsidies divided by the number of members, i.e. cost per membership. This is at least a good starting point in estimating costs.

However, we have to look for an estimationof the output in our example. Because, as we mentioned, we are interested in social costs, output can be a social welfare parameter like unemployment. We can compare employment development in regions with microfinance services abd regions without it, estimating the gain employment. Therefore, as a result, we estimate the costs per created job.
